```python
import pandas as pd
from fuzzywuzzy import fuzz
```

```python
# Using pandas to read csv files into Dataframes.
import_sheet = pd.read_csv('import_data.csv')  
data_bank_sheet = pd.read_csv('bank_data.csv')
```

# TASK 1

Task 1 â€“ Return Brand Name and Model Name in import table from Data bank table where product description of import file is 100% match with product description of Data Bank. 

```python

# Function to calculate the percentage match between two product descriptions.
def calculate_match(desc1, desc2):
    return fuzz.token_set_ratio(desc1.lower(), desc2.lower())

# Create an empty list to store matching rows.
matching_rows = []

# Iterate through each row in the 'import_sheet' dataframe.
for import_index, import_row in import_sheet.iterrows():
    max_match = 0
    best_match_brand = ""
    best_match_model = ""
    
    # Iterate through each row in the 'data_bank_sheet' dataframe to find the best match.
    for data_bank_index, data_bank_row in data_bank_sheet.iterrows():
        match_percentage = calculate_match(import_row['Product Description'], data_bank_row['Product Description'])
        if match_percentage > max_match:
            max_match = match_percentage
            best_match_brand = data_bank_row['Brand Name']
            best_match_model = data_bank_row['Model Name']
    
    # If a 100% match is found, add the brand and model names to the matching_rows list.
    if max_match == 100:
        matching_rows.append({
            'Product Description': import_row['Product Description'],
            'Model Name': best_match_model,
            'Brand Name': best_match_brand,
            'Match Percentage': max_match
        })

# Create a new dataframe using the matching_rows list.
result_df = pd.DataFrame(matching_rows)

# Now, you can merge the 'import_sheet' with 'result_df' based on 'Product Description'.
# This will add the 'Model Name', 'Brand Name', and 'Match Percentage' columns to 'import_sheet'.
import_sheet = import_sheet.merge(result_df, on='Product Description', how='left')

# Print the updated 'import_sheet' dataframe.
print(import_sheet)
```

# TASK 2

Task 2 - Return Brand Name and Model name column in import table from Data bank table where product description of import file is up to 90% match with product description of Data Bank. (Consider only those product descriptions of import table for which Brand Name and Model name column is not returned in task 1).

```python

# Function to calculate the percentage match between two product descriptions.
def calculate_match(desc1, desc2):
    return fuzz.token_set_ratio(desc1.lower(), desc2.lower())

# Create a new dataframe to store the matching results.
matching_rows = []

# Iterate through each row in the 'import_sheet' dataframe.
for import_index, import_row in import_sheet.iterrows():
    max_match = 0
    best_match_brand = ""
    best_match_model = ""
    
    # Iterate through each row in the 'data_bank_sheet' dataframe to find the best match.
    for data_bank_index, data_bank_row in data_bank_sheet.iterrows():
        match_percentage = calculate_match(import_row['Product Description'], data_bank_row['Product Description'])
        if match_percentage > max_match and match_percentage <= 90:
            max_match = match_percentage
            best_match_brand = data_bank_row['Brand Name']
            best_match_model = data_bank_row['Model Name']
    
    # If a match up to 90% is found, add the brand and model names to the matching_rows list.
    if max_match > 0:
        matching_rows.append({
            'Product Description': import_row['Product Description'],
            'Model Name': best_match_model,
            'Brand Name': best_match_brand,
            'Match Percentage': max_match
        })

# Create a new dataframe using the matching_rows list.
result_df = pd.DataFrame(matching_rows)

# Now, you can merge the 'import_sheet' with 'result_df' based on 'Product Description'.
# This will add the 'Model Name', 'Brand Name', and 'Match Percentage' columns to 'import_sheet'.
import_sheet = import_sheet.merge(result_df, on='Product Description', how='left')

# Print the updated 'import_sheet' dataframe with the new matches.
print(import_sheet)
```
